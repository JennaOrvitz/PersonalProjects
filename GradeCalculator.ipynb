```python
semester_grades={
    'First year fall':{'Math 1X03':9, 'Math 1B03':8, 'Math 1MP3':12},
    'First year winter':{'Compsci 1DM3':12, 'Compsci 1XC3':8, 'Innovate 1Z03':12, 'Math 1XA3':11, 'Math 1XX3':8},

    'Second year fall':{'Math 2LA3': 11, 'Math 2X03': 9, 'Music 2MT3':12, 'Stats 2D03':12},
    'Second year winter':{'Math 2C03':11, 'Math 2XX3':10, 'Music 2MU3': 12, 'Stats 2DA3': 12, 'Stats 2MB3': 12},

    'Third year fall':{'Econ 2P03':10, 'Math 3NA3':12, 'Stats 3A03':11, 'Stats 3D03':9, 'Stats 3PG3': 11},
    #'Third_ year winter':{'Datasci 3ML3':'', 'Math 3IA3':'', 'Math 3MB3':'', 'Stats 3DA3':''}

    #'Fourth year fall':{}
    #'Fourth year winter':{}
}
overall_grades = [grade for semester_grades in semester_grades.values() for grade in semester_grades.values()]
overall_average = sum(overall_grades) / len(overall_grades)

print(semester_grades)

print(f'Overall Average: {overall_average:.4f}')
```

```python
import matplotlib.pyplot as plt
semester_averages = {}
for semester, grades in semester_grades.items():
    average = sum(grades.values()) / len(grades)
    semester_averages[semester] = average

# Plotting
plt.plot(semester_averages.keys(), semester_averages.values(), marker='o', linestyle='-')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Semester')
plt.ylabel('Average Grade')
plt.title('Semester Averages (12.0 Scale)')
plt.ylim(9, 12) 
plt.show()
```

```python
conversion_mapping = [
    (12.0, 4.0),
    (11.0, 3.9),
    (10.0, 3.7),
    (9.0, 3.3),
    (8.0, 3.0)
]

semester_grades_4_scale = {
    semester: {course: next((value for key, value in conversion_mapping if key == grade), grade) for course, grade in grades.items()}
    for semester, grades in semester_grades.items()
}

print(semester_grades_4_scale)

overall_grades = [grade for semester_grades in semester_grades_4_scale.values() for grade in semester_grades.values()]
overall_average = sum(overall_grades) / len(overall_grades)

print(f'Overall Average: {overall_average:.4f}')
```

```python
import matplotlib.pyplot as plt
semester_averages_4 = {}
for semester, grades in semester_grades_4_scale.items():
    average = sum(grades.values()) / len(grades)
    semester_averages_4[semester] = average

# Plotting
plt.plot(semester_averages_4.keys(), semester_averages_4.values(), marker='o', linestyle='-')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Semester')
plt.ylabel('Average Grade')
plt.title('Semester Averages (4.0 Scale)')
plt.ylim(3.2, 4) 
plt.show()
```
